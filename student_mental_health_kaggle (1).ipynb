# Data Visualization

```python
plt.figure(figsize=(6,6))
plt.hist(df['Age'],color='c')
plt.title("Age distribution");
```

```python
df.rename(columns = {'Choose your gender': 'gender'}, inplace = True)
```

```python
plt.figure(figsize=(12,6))
plt.title("gender distribution")
g = plt.pie(df.gender.value_counts(), explode=(0.025,0.025), labels=df.gender.value_counts().index, colors=['skyblue','navajowhite'],autopct='%1.1f%%', startangle=180);
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(6,6))
plt.hist(df['Your current year of Study'],color='r')
plt.title("Students studying in particular year");
```

```python
plt.figure(figsize=(6,6))
ax = sns.countplot(x='Your current year of Study', hue="gender", data=df)
plt.title("Students studying in particular year")
plt.show()
```

```python
plt.figure(figsize=(8,8))
ax = sns.countplot(x='Do you have Anxiety?', hue="Do you have Depression?", data=df)
plt.title("Students studying in particular year")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="darkgrid")
ax = sns.countplot(y="Do you have Anxiety?", hue="gender", data=df)
plt.title("Anxiety by Gender")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="darkgrid")
ax = sns.countplot(y="Do you have Depression?", hue="gender", data=df)
plt.title("Depression by Gender")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Do you have Anxiety?", hue="Age", data=df)
plt.title("Anxiety by age")
plt.show()
```

```python
plt.figure(figsize=(30,15))
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Do you have Panic attack?", hue="What is your course?", data=df)
plt.title("Panick attack by course")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Do you have Panic attack?", hue="What is your CGPA?", data=df)
plt.title("Panic attack by CGPA")
plt.show()
```

```python
plt.figure(figsize=(8,8))
ax = sns.countplot(x='Did you seek any specialist for a treatment?', hue="gender", data=df)
plt.title("Asking for help by gender")
plt.show()
```

```python
plt.figure(figsize=(6,6))
ax = sns.countplot(x='Do you have Depression?', hue="Marital status", data=df)
plt.title("Depression depending on martial status")
plt.show()
```
